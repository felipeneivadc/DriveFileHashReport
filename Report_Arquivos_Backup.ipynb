```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

# Caminho para o relatório gerado anteriormente
input_report_path = '/content/drive/My Drive/Relatorio_Arquivos_Felipe_Neiva.xlsx'

# Caminho para o novo relatório de exclusão
output_deletion_report_path = '/content/drive/My Drive/Relatorio_Exclusao_Arquivos_Felipe_Neiva.xlsx'

# Ler o relatório original
df = pd.read_excel(input_report_path)

# Lista para armazenar os dados do relatório de exclusão
deletion_data = []

# Função para tentar excluir um arquivo
def try_delete_file(file_path):
    try:
        os.remove(file_path)
        return "Excluído com sucesso", None
    except FileNotFoundError:
        return None, f"Erro: Arquivo não encontrado para exclusão: {file_path}"
    except PermissionError:
        return None, f"Erro: Permissão negada ao tentar excluir o arquivo: {file_path}"
    except Exception as e:
        return None, f"Erro desconhecido ao tentar excluir o arquivo {file_path}. Erro: {e}"

# Contador para monitorar progresso
file_counter = 0
total_files_to_delete = len(df[df['Excluir Arquivo'] == 'Sim'])

print(f"Total de arquivos a serem excluídos: {total_files_to_delete}")

# Processar cada arquivo no DataFrame
for index, row in df.iterrows():
    if row['Excluir Arquivo'] == 'Sim':
        file_counter += 1
        file_path = row['Caminho Completo']
        print(f"Excluindo arquivo {file_counter}/{total_files_to_delete}: {row['Nome do Arquivo']}")

        status, error_message = try_delete_file(file_path)

        if error_message:
            deletion_data.append({
                "Caminho Completo": file_path,
                "Nome do Arquivo": row['Nome do Arquivo'],
                "Status de Exclusão": "Falhou",
                "Erro Apresentado": error_message
            })
            print(f"Erro ao excluir {file_path}: {error_message}")
        else:
            deletion_data.append({
                "Caminho Completo": file_path,
                "Nome do Arquivo": row['Nome do Arquivo'],
                "Status de Exclusão": "Sucesso",
                "Erro Apresentado": "Nenhum"
            })
            print(f"Arquivo excluído com sucesso: {file_path}")

# Cria um DataFrame com os dados da exclusão
deletion_df = pd.DataFrame(deletion_data)

print("Exclusão de arquivos concluída. Gerando relatório...")

# Salva o DataFrame em um arquivo Excel (Relatório de exclusão)
deletion_df.to_excel(output_deletion_report_path, index=False)

print(f'Relatório de exclusão gerado com sucesso: {output_deletion_report_path}')
```
